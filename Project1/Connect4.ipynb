```python
# IMPORTING THE DATA
import pandas as pd
```

```python
data =  pd.read_csv('connectfour.data', sep=",")
```

```python
data.head()
```

```python
data.shape
```

```python
# MAKING DATA SUITABLE FOR MODEL
data.replace('b', 0, inplace = True)
data.replace('x', 1, inplace = True)
data.replace('o', -1, inplace = True)
data.head()
```

```python
data.replace('win', 0, inplace = True)
data.replace('loss', 1, inplace = True)
data.replace('draw', 2,inplace = True)
```

```python
data.head()
```

```python
data['win']
```

```python
# SPLITING THE DATA IN TRAIN AND TEST
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns = "win"), data['win'], test_size = 0.20, shuffle = True, random_state = 1024)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.25, shuffle = True, random_state = 1024)
```

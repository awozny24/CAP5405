```python
# IMPORTING THE DATA
import pandas as pd
```

```python
colnames = ['a1','a2','a3','a4','a5','a6','b1','b2','b3','b4','b5','b6','c1','c2','c3','c4','c5','c6','d1','d2','d3','d4','d5','d6','e1','e2','e3','e4','e5','e6','f1','f2','f3','f4','f5','f6','g1','g2','g3','g4','g5','g6','result']
data =  pd.read_csv('/content/drive/MyDrive/DLCG_Project/connectfour.data', names=colnames,sep="," )
```

```python
# MAKING DATA SUITABLE FOR MODEL
data.replace('b', 0, inplace = True)
data.replace('x', 1, inplace = True)
data.replace('o', -1, inplace = True)
data
# data.iloc[0]
```

```python
data.replace('win', 0, inplace = True)
data.replace('loss', 1, inplace = True)
data.replace('draw', 2,inplace = True)
```

```python
# SPLITING THE DATA IN TRAIN AND TEST
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns = "result"), data['result'], test_size = 0.1, shuffle = True, random_state = 1024)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.2, shuffle = True, random_state = 1024)
```

```python
X_val.shape, y_val.shape
```

```python
# TENSORFLOW IMPORTS
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.utils import to_categorical
```

```python
y_train_oh = to_categorical(y_train, num_classes=3)
y_val_oh = to_categorical(y_val, num_classes=3)
y_test_oh = to_categorical(y_test, num_classes=3)
```

```python
y_train_oh
```

```python
epochs = 30
batch_size = 300
```

```python
# CREATING THE MODEL
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import Input

model = Sequential()
model.add(Input(shape = (X_train.shape[1])))
# model.add(Dense(512, activation='relu'))
# model.add(Dense(256, activation='relu'))
# model.add(Dense(128, activation='relu'))
# model.add(Dense(64, activation='relu'))
model.add(Dense(43, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax',kernel_regularizer=tf.keras.regularizers.L1(0.01), name='Output'))

model.summary()
```

```python
# COMPILING THE MODEL
loss = tf.keras.losses.CategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss=loss, metrics= ['accuracy'])
```

```python
# REDUCING LEARNING RATE ON PLATEAU
from tensorflow.keras.callbacks import ReduceLROnPlateau

lrr = ReduceLROnPlateau(monitor='val_accuracy', 
                        patience=5, 
                        verbose=1, 
                        factor=0.4, 
                        min_lr=0.0001)
```

```python
# MODEL TRAINING
history = model.fit(X_train, y_train_oh, 
                    validation_data=(X_test, y_test_oh), 
                    epochs=epochs, 
                    batch_size=batch_size,
                    steps_per_epoch=X_train.shape[0]//batch_size,
                    validation_steps=X_test.shape[0]//batch_size,
                    callbacks=[lrr],
                    shuffle=True)
```

```python
# SAVING THE MODEL
model.save("/content/drive/MyDrive/DLCG_Project/model.h5")
```

```python
# PLOTTING THE ACCURACY AND VALIDATION ACCURACY VS EPOCHS
from matplotlib import pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
model.evaluate(X_val, y_val_oh)
```
